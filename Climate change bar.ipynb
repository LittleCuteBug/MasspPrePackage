```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("climate-change-2016.csv")
```

```python
df.info()
```

```python
np.asarray(df["CO2(ppm)"][::5])
```

```python
#plt.plot(df.year[::5],df.DJIA[::5],"bo")
#plt.plot(df.year[::5],np.asarray(df["CO2(ppm)"][::5]),"gx")
plt.figure(figsize = (10,10))
plt.plot("year","DJIA",data = df[::5])
plt.plot("year","CO2(ppm)",data = df[::5])
plt.axis([1950,2020,0,20000])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.bar("year","DJIA", 1,data = df[::2])
plt.xlabel("A")

plt.subplot(1,2,2)
plt.plot("year","CO2(ppm)",data = df[::2],c = "r",marker = 'o')
plt.xlabel("B")
```
